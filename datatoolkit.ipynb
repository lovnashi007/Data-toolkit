###2. How does broadcasting work in Numpy?
- Broadcasting in Numpy is a technique that allows operations between arrays of different shapes without manually reshaping them. NumPy automatically expands the smaller array along the required dimensions so that the operation can be performed element-wise.

```python

A = np.array([[1, 2, 3],
              [4, 5, 6]])   # Shape (2, 3)
B = [1,2,3]  # Shape (3,)

C = A + B
print(C)


```

###13. Why is handling missing data important in Pandas?

###17. Explain difference between apply() and map() in pandas.

- map() is used only with Series(single column) to apply a function element-wise.


- apply() can be used with both DataFrame and Series to apply a function along an axis(rows or columns).

```phton

import pandas as pd

# Using map() on Series
s = pd.Series([1, 2, 3])
print(s.map(lambda x: x**2))   # [1, 4, 9]

# Using apply() on DataFrame
df = pd.DataFrame({'A':[1, 2, 3], 'B':[4, 5, 6]})
print(df.apply(sum))   # Sum of each column


```

###20. What is the role of pivot table in Python ?
-  In Python, especially using Pandas, a pivot table is used to summarize and analyze data efficiently. it allows you to reorganize data by specifying rows, columns, and values to aggregate(like sum, mean, count,etc.) you can create one using pd.pivot_table(). pivot table help identify patterns and insights by grouping data and performing calculations automatically. For example, they can show total sales by product and region, maiking large datasets easier to interpret and compare.


```python

import pandas as pd

# Sample data
data = {
    'Product': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Region': ['North', 'South', 'North', 'South', 'North', 'South'],
    'Sales': [100, 150, 200, 250, 300, 350]
}

df = pd.DataFrame(data)

# Create pivot table
pivot = pd.pivot_table(df,
                       values='Sales',      # What to aggregate
                       index='Product',     # Rows
                       columns='Region',    # Columns
                       aggfunc='sum',       # Aggregation function
                       fill_value=0)        # Fill missing values

print(pivot)


```

###21. Why is NumPy's array slicing faster than Python's list slicing?

- NumPy arrays are stored in contiguous blocks of memory, unlike python lists, which store references to separate objects. This allows NumPy to perform slicing without copying data, just creating a view of the original array.Python lists, on the other hand, require creating a new list and copying elemnts for every slice, which is slower.Additionally, NumPy is implemented in C, so operations on arrays(including slicing)  are vectorized and optimized, making them much faster than python's built-in-list operations.
